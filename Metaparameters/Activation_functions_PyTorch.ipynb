```python
import torch
import matplotlib.pyplot as plt
plt.rcParams.update({'font.size':18})
```

```python
x = torch.linspace(-3,3,101)

def NNoutputx(act_func):
    act_func = getattr(torch, act_func)
    return act_func(x)
```

```python
activation_funcs = ['relu', 'sigmoid', 'tanh']

fig = plt.figure(figsize=(10, 8))

for act_func in activation_funcs:
    plt.plot(x, NNoutputx(act_func), label=act_func, linewidth=3)
    
dashlinecolor=[.7, .7, .7]
plt.plot(x[[0, -1]], [0, 0], '--', color=dashlinecolor)
plt.plot(x[[0, -1]], [1, 1], '--', color=dashlinecolor)
plt.plot([0, 0], [-1, 3], '--', color=dashlinecolor)

plt.legend()
plt.xlabel('x')
plt.ylabel('$\sigma(x)$')
plt.title('Various activation fuctions')
plt.xlim(x[[0,-1]])
plt.ylim([-1,3])

plt.show()
```

```python
def NNoutputx(act_func):
    act_func = getattr(torch.nn, act_func)
    return act_func()
```

```python
activation_funcs= ['ReLU6', 'Hardshrink', 'LeakyReLU']

fig = plt.figure(figsize=(10, 8))

for act_func in activation_funcs:
    plt.plot(x, NNoutputx(act_func)(x), label=act_func, linewidth=3)
    
dashlinecolor=[.7, .7, .7]
plt.plot(x[[0, -1]], [0, 0], '--', color=dashlinecolor)
plt.plot(x[[0, -1]], [1, 1], '--', color=dashlinecolor)
plt.plot([0, 0], [-1, 3], '--', color=dashlinecolor)

plt.legend()
plt.xlabel('x')
plt.ylabel('$\sigma(x)$')
plt.title('Various activation fuctions')
# plt.xlim(x[[0,-1]])
# plt.ylim([-.1,.1])

plt.show()
```

```python
x = torch.linspace(-3, 9, 101)
relu6 = torch.nn.ReLU6()
plt.plot(x, relu6(x))
plt.show()
```

```python
x = torch.linspace(-3, 3, 21)

# in torch
y1 = torch.relu(x)

# in torch.nn
f = torch.nn.ReLU()
y2 = f(x)

plt.plot(x,y1,'ro',label='torch.relu')
plt.plot(x,y2,'bx',label='torch.nn.ReLU()')
plt.legend()
plt.xlabel('Input')
plt.ylabel('Output')
plt.show()
```

```python
x1 = torch.linspace(-1,1,20)
x2 = x1**2

w1 = .6
w2 = .8

linpart = x1*w1+x2*w2

y = torch.relu(linpart)

plt.plot(x1,linpart,'bo-',label='Linear input')
plt.plot(x1,y,'rs',label='Nonlinear output')
plt.ylabel("$\\hat{y}$ (output of activation function)")
plt.xlabel('x1 variable')
# plt.ylabel([-.1, .1])
plt.legend()
plt.show()
```
