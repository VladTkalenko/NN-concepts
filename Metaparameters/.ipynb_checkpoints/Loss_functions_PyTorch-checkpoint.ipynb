```python
import torch
import torch.nn as nn

import numpy as np
import matplotlib.pyplot as plt
```

# MSELoss

```python
loss_fun_mse = nn.MSELoss()

yHat = torch.linspace(-2,2,101)
y = torch.tensor(.5)

losses = np.zeros(101)
for i, yy in enumerate(yHat):
    losses[i] = loss_fun_mse(yy, y)
    
plt.plot(yHat, losses, label='Loss')
plt.plot([y,y], [0, np.max(losses)], 'r--', label='True value')
plt.xlabel('Predicted value')
plt.legend()
plt.show()
```

# Binary Cross-entropy

```python
loss_fun_bce = nn.BCELoss()

yHat = torch.linspace(.001, .999, 101)
y1 = torch.tensor(0.)
y2 = torch.tensor(1.)

losses = np.zeros((101, 2))
for i, yy in enumerate(yHat):
    losses[i,0] = loss_fun_bce(yy,y1)
    losses[i,1] = loss_fun_bce(yy,y2)
    
plt.plot(yHat, losses)
plt.xlabel('Predicted Value')
plt.ylabel('Loss')
plt.legend(['correct=0', 'correct=1'])
plt.show()
```

# Categorical cross-entropy

```python
loss_func_cce = nn.CrossEntropyLoss()

yHat = torch.tensor([[1., 4, 3]])

for i in range(3):
    correct_answer = torch.tensor([i])
    this_loss = loss_func_cce(yHat, correct_answer).item()
    print(f"Loss when corrected answer is {i}: {this_loss}")
```

```python
# THIS IS INCORRECT!!!!!
sm = nn.Softmax(dim=1)
yHat_sm = sm(yHat)

for i in range(3):
    correct_answer = torch.tensor([i])
    this_loss = loss_func_cce(yHat_sm, correct_answer).item()
    print(f"Loss when corrected answer is {i}: {this_loss}")
```

```python
# compare raw, softmax, and log-softmax output
sm = nn.LogSoftmax(dim=1)
yHat_logsm = sm(yHat)
print(yHat)
print(yHat_sm)
print(yHat_logsm)
```

# Creating custom loss function

```python
class MyLoss(nn.Module):
    def __init__(self):
        super().__init__()
        
    def forward(self,x,y):
        loss = torch.abs(x-y)
        return loss
    
loss_func = MyLoss()
loss_func(torch.tensor(4), torch.tensor(5.2))
```
