```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import TensorDataset, DataLoader
from sklearn.model_selection import train_test_split
import time
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import scipy.stats as stats

import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"

data = pd.read_csv(url, sep=';')

data = data[data['total sulfur dioxide']<200]

cols_to_z_score = data.keys()
cols_to_z_score = cols_to_z_score.drop('quality')

# for col in cols_to_z_score:
#     meanval = np.mean(data[col])
#     stdev = np.std(data[col], ddof=0)
#     data[col] = (data[col]-meanval) / stdev
data[cols_to_z_score] = data[cols_to_z_score].apply(stats.zscore)   

data['bool_quality'] = 0
data['bool_quality'][data['quality']>5] = 1
```

```python
dataT = torch.tensor(data[cols_to_z_score].values).float()
labels = torch.tensor(data['bool_quality'].values).float()

labels = labels[:, None]

train_data, test_data, train_labels, test_labels = train_test_split(dataT, labels, test_size=.1)

train_dataset = TensorDataset(train_data, train_labels)
test_dataset = TensorDataset(test_data, test_labels)

batch_size=32
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, drop_last=True)
test_loader = DataLoader(test_dataset, batch_size=test_dataset.tensors[0].shape[0])
```

```python
class ANNWine(nn.Module):
    def __init__(self):
        super().__init__()
        
        self.input = nn.Linear(11, 16)
        
        self.fc1 = nn.Linear(16, 32)
        self.fc2 = nn.Linear(32, 32)
        
        self.output = nn.Linear(32, 1)
        
        
    def forward(self, x):
        
        x = F.relu(self.input(x))
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
          
        return self.output(x)
```

```python
num_of_epochs = 200

def train_model(model):
    loss_func = nn.BCEWithLogitsLoss()
    optimizer = torch.optim.SGD(model.parameters(), lr=.01)
    
    train_acc = []
    test_acc = [] 
    losses = torch.zeros(num_of_epochs)
    
    for epoch in range(num_of_epochs):
        model.train()
        batch_acc = []
        batch_loss = []
        for X, y in train_loader:
            yHat = model(X)
            loss = loss_func(yHat, y)
            
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            predlabels = yHat>0
            batch_acc.append(100*torch.mean((predlabels==y).float()).item())
            batch_loss.append(loss.item())
            
        train_acc.append(np.mean(batch_acc))
        losses[epoch] = np.mean(batch_loss)
        
        model.eval()
        X,y = next(iter(test_loader))
        with torch.no_grad():
            predlabels = model(X)>0
        test_acc.append(100*torch.mean((predlabels==y).float()).item())
    
    return train_acc, test_acc, losses
```

```python
model_xavier = ANNWine()
for p in model_xavier.named_parameters():
    if 'bias' in p[0]:
        nn.init.xavier_normal_(p[1].data.view(len(p[1].data), 1))
    elif 'weight' in p[0]:
        nn.init.xavier_normal_(p[1].data)
```

```python
model_kaiming = ANNWine()
for p in model_kaiming.named_parameters():
    if 'bias' in p[0]:
        nn.init.kaiming_uniform_(p[1].data.view(len(p[1].data), 1))
    elif 'weight' in p[0]:
        nn.init.kaiming_uniform_(p[1].data)
```

```python
all_weights_xavier = np.array([])
all_biases_xavier = np.array([])
all_weights_kaiming = np.array([])
all_biases_kaiming = np.array([])

for p in model_xavier.named_parameters():
    if 'bias' in p[0]:
        all_biases_xavier = np.concatenate((all_biases_xavier, p[1].data.numpy().flatten()), axis=0)
    elif 'weight' in p[0]:
        all_weights_xavier = np.concatenate((all_weights_xavier, p[1].data.numpy().flatten()), axis=0)   
        
for p in model_kaiming.named_parameters():
    if 'bias' in p[0]:
        all_biases_kaiming = np.concatenate((all_biases_kaiming, p[1].data.numpy().flatten()), axis=0)
    elif 'weight' in p[0]:
        all_weights_kaiming = np.concatenate((all_weights_kaiming, p[1].data.numpy().flatten()), axis=0)   
    
    

fig, ax = plt.subplots(2,2,figsize=(12,12))

ax[0, 0].hist(all_biases_xavier, 40)
ax[0, 0].set_title('Histogram of initial biases gaussian')

ax[0, 1].hist(all_weights_xavier, 40)
ax[0, 1].set_title('Histogram of initial weights gaussian')

ax[1, 0].hist(all_biases_kaiming, 40)
ax[1, 0].set_title('Histogram of initial biases uniform')

ax[1, 1].hist(all_weights_kaiming, 40)
ax[1, 1].set_title('Histogram of initial weights uniform')


for item in ax.flatten():
    item.set_xlabel('Initial value')
    item.set_ylabel('Count')
```

```python
train_acc_xavier, test_acc_xavier, losses_xavier = train_model(model_xavier)
train_acc_kaiming, test_acc_kaiming, losses_kaiming = train_model(model_kaiming)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].plot(losses_xavier)
ax[0].plot(losses_kaiming)
ax[0].set_ylabel('Loss')
ax[0].set_title('Losses')
ax[0].set_xlabel('Epoch')

ax[1].plot(test_acc_xavier)
ax[1].plot(test_acc_kaiming)
ax[1].set_ylabel('Accuracy (%)')
ax[1].set_title('Test accuracies')
ax[1].set_xlabel('Epoch')

plt.legend(['Xavier', 'Kaiming'])
plt.show()
```

```python
test_accuracies_xavier = []
test_accuracies_kaiming = []

loss_10_xavier = []
loss_10_kaiming = []

train_accuracies_xavier = []
train_accuracies_kaiming = []


for i in range(50):
    model_xavier = ANNWine()
    for p in model_xavier.named_parameters():
        if 'bias' in p[0]:
            nn.init.xavier_normal_(p[1].data.view(len(p[1].data), 1))
        elif 'weight' in p[0]:
            nn.init.xavier_normal_(p[1].data)
            
    model_kaiming = ANNWine()
    for p in model_kaiming.named_parameters():
        if 'bias' in p[0]:
            nn.init.kaiming_uniform_(p[1].data.view(len(p[1].data), 1))
        elif 'weight' in p[0]:
            nn.init.kaiming_uniform_(p[1].data)  
            
    train_acc_xavier, test_acc_xavier, losses_xavier = train_model(model_xavier)
    train_acc_kaiming, test_acc_kaiming, losses_kaiming = train_model(model_kaiming)
    
    test_accuracies_xavier.append(np.mean(test_acc_xavier[-5:]))
    test_accuracies_kaiming.append(np.mean(test_acc_kaiming[-5:]))
    
    train_accuracies_xavier.append(np.mean(train_acc_xavier[-5:]))
    train_accuracies_kaiming.append(np.mean(train_acc_kaiming[-5:]))
    
    loss_10_xavier.append(torch.mean(losses_xavier[-5:]).item())
    loss_10_kaiming.append(torch.mean(losses_kaiming[-5:]).item())
```

```python
t_losses, p_losses = stats.ttest_ind(loss_10_xavier, loss_10_kaiming)
t_train_acc, p_train_acc = stats.ttest_ind(train_accuracies_xavier, train_accuracies_kaiming)
t_test_acc, p_test_acc = stats.ttest_ind(test_accuracies_xavier, test_accuracies_kaiming)
```

```python
fig, ax = plt.subplots(1, 3, figsize=(12, 6))

ax[0].plot(np.zeros(50), loss_10_xavier, 'o')
ax[0].plot(np.ones(50), loss_10_kaiming, 'o')
ax[0].set_xlim([-1, 2])
ax[0].set_xticks([0, 1])
ax[0].set_xticklabels(["Xavier", "Kaiming"])
ax[0].set_title(f"Losses: t={round(t_losses, 2)}, p={round(p_losses, 2)}")

ax[1].plot(np.zeros(50), train_accuracies_xavier, 'o')
ax[1].plot(np.ones(50), train_accuracies_kaiming, 'o')
ax[1].set_xlim([-1, 2])
ax[1].set_xticks([0, 1])
ax[1].set_xticklabels(["Xavier", "Kaiming"])
ax[1].set_title(f"Train accuracy: t={round(t_train_acc, 2)}, p={round(p_train_acc, 2)}")

ax[2].plot(np.zeros(50), test_accuracies_xavier, 'o')
ax[2].plot(np.ones(50), test_accuracies_kaiming, 'o')
ax[2].set_xlim([-1, 2])
ax[2].set_xticks([0, 1])
ax[2].set_xticklabels(["Xavier", "Kaiming"])
ax[2].set_title(f"Test accuracy: t={round(t_test_acc, 2)}, p={round(p_test_acc, 2)}")
```
