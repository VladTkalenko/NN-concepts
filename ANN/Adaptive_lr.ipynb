```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
import matplotlib_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
```

```python
def fx(x):
    return 3*x**2-3*x+4

def deriv(x):
    return 6*x-3
```

```python
x = np.linspace(-2, 2, 2001)

plt.plot(x,fx(x), x,deriv(x))
plt.xlim(x[0], x[-1])
plt.grid()
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(['y', 'dy'])
plt.show()
```

```python
localmin = np.random.choice(x, 1)

learning_rate = 0.01
training_epochs = 100

model_params_fixed = np.zeros((training_epochs, 3))
for i in range(training_epochs):
    grad = deriv(localmin)
    lr = learning_rate
    localmin = localmin - learning_rate*grad
    
    model_params_fixed[i, :] = localmin, grad, lr
localmin
```

```python
localmin = np.random.choice(x, 1)
initval = localmin[:]

model_params_grad = np.zeros((training_epochs, 3))
for i in range(training_epochs):
    grad = deriv(localmin)
    lr = learning_rate*np.abs(grad)
    localmin = localmin - learning_rate*grad
    
    model_params_grad[i, :] = localmin, grad, lr
localmin
```

```python
learning_rate = .1
localmin = initval

model_params_time = np.zeros((training_epochs, 3))
for i in range(training_epochs):
    grad = deriv(localmin)
    lr = learning_rate*(1-(i+1)/training_epochs)
    localmin = localmin - learning_rate*grad
    
    model_params_time[i, :] = localmin, grad, lr
localmin
```

```python
fig, ax = plt.subplots(1,3, figsize=(10,3))

for i in range(3):
    ax[i].plot(model_params_fixed[:, i], 'o-', markerfacecolor='w')
    ax[i].plot(model_params_grad[:, i], 'o-', markerfacecolor='w')
    ax[i].plot(model_params_time[:, i], 'o-', markerfacecolor='w')
    ax[i].set_xlabel('Iteration')
    
ax[0].set_ylabel('Local minimum')
ax[1].set_ylabel('Derivative')
ax[2].set_ylabel('Learning rate')
ax[2].legend(['Fixed l.r.', 'Grad-based l.r.', 'Time-based l.r.'])

plt.tight_layout()
plt.show()
```
