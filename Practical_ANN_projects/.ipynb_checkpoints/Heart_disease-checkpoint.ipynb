```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from scipy import stats

import torch
import torch.nn as nn
from torch.utils.data import TensorDataset, DataLoader

from sklearn.model_selection import train_test_split
```

```python
# removes rows with outliers:
data_without_outliers = data_frame[data_frame['chol']<500]

fig, ax = plt.subplots(1, figsize=(12, 5))
ax = sns.boxplot(data=data_without_outliers)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
plt.show()
```

```python
cols_to_z_score = data_without_outliers.keys()
cols_to_z_score = cols_to_z_score.drop('target')

# for col in cols_to_z_score:
#     meanval = np.mean(data[col])
#     stdev = np.std(data[col], ddof=0)
#     data[col] = (data[col]-meanval) / stdev
data_without_outliers[cols_to_z_score] = data_without_outliers[cols_to_z_score].apply(stats.zscore)    
data_without_outliers.describe()
```

```python
fig, ax = plt.subplots(1, figsize=(17,4))
ax = sns.boxplot(data=data_without_outliers)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)
plt.show()
```

```python
dataT = torch.tensor(data_without_outliers[cols_to_z_score].values).float()
labels = torch.tensor(data_without_outliers['target'].values).float()

print(dataT.shape)
print(labels.shape)

labels = labels[:, None]
print(labels.shape)
```

```python
values, counts = torch.unique(labels, return_counts=True)
print(f"Unique labels: {values}, count: {counts}")
```

```python
train_data, test_data, train_labels, test_labels = train_test_split(dataT, labels, test_size=.1)

train_dataset = TensorDataset(train_data, train_labels)
test_dataset = TensorDataset(test_data, test_labels)

batch_size=32
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, drop_last=True)
test_loader = DataLoader(test_dataset, batch_size=test_dataset.tensors[0].shape[0])
```

```python
def create_model():
    Disease_model = nn.Sequential(
    nn.Linear(13, 16),
    nn.ReLU(),
    nn.Linear(16, 32),
    nn.ReLU(),
    nn.Linear(32, 32),
    nn.ReLU(),
    nn.Linear(32, 1),
    )
    
    loss_fun = nn.BCEWithLogitsLoss()
    optimizer = torch.optim.Adam(Disease_model.parameters(), lr=.01)
    
    return Disease_model, loss_fun, optimizer
```

```python
num_of_epochs = 100

def train_model(model, loss_func, optimizer):
    best_model = {
        "Accuracy": 0,
        "Epoch": 0,
        "Model_state": None
    }
    
    train_acc = []
    test_acc = [] 
    losses = []
    
    for epoch in range(num_of_epochs):
        model.train()
        batch_acc = []
        batch_loss = []
        for X, y in train_loader:
            yHat = model(X)
            loss = loss_func(yHat, y)
            
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            predlabels = yHat>0
            batch_acc.append(100*torch.mean((predlabels==y).float()).item())
            batch_loss.append(loss.item())
            
        train_acc.append(np.mean(batch_acc))
        losses.append(np.mean(batch_loss))
        
        model.eval()
        X,y = next(iter(test_loader))
        predlabels = model(X)>0
        test_acc.append(100*torch.mean((predlabels==y).float()).item())
        
        if test_acc[-1] > best_model["Accuracy"]:
            best_model["Accuracy"] = test_acc[-1]
            best_model["Model_state"] = model.state_dict()
            best_model["Epoch"] = epoch
    
    return train_acc, test_acc, losses, best_model
```

```python
model, loss_func, optimizer = create_model()

train_acc, test_acc, losses, best_model = train_model(model, loss_func, optimizer)
```

```python
print(f"The best accuracy: {best_model['Accuracy']}% on {best_model['Epoch']} epoch")
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 5))

ax[0].plot(losses)
ax[0].set_xlabel('Epoch')
ax[0].set_ylabel('Loss')
ax[0].set_title("Model loss")

ax[1].plot(train_acc, label='Train')
ax[1].plot(test_acc, label='Test')
ax[1].set_xlabel('Epoch')
ax[1].set_ylabel('Accuracy (%)')
ax[1].set_title('Model accuracy')
ax[1].set_ylim([80, 102])
ax[1].legend()

plt.show()
```

```python
model2 = create_model()[0]
model2.load_state_dict(best_model['Model_state'])
```

```python
def predict(item_number):
    item = test_dataset.tensors[0][item_number, :]
    label = test_dataset.tensors[1][item_number].item()

    prediction =  (model2(item)>0).float().item()
    print(f'Real state: {label}, predicted: {prediction}')
```
