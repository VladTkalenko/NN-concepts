```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import TensorDataset, DataLoader
from sklearn.model_selection import train_test_split
import time
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import scipy.stats as stats

import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"

data = pd.read_csv(url, sep=';')

data = data[data['total sulfur dioxide']<200]

cols_to_z_score = data.keys()
cols_to_z_score = cols_to_z_score.drop('quality')

# for col in cols_to_z_score:
#     meanval = np.mean(data[col])
#     stdev = np.std(data[col], ddof=0)
#     data[col] = (data[col]-meanval) / stdev
data[cols_to_z_score] = data[cols_to_z_score].apply(stats.zscore)  

```

```python
def binarize_quality(threshold):

    data['bool_quality'] = 0
    data['bool_quality'][data['quality']>threshold] = 1
    
    return data

def get_train_test_data(threshold):
    data = binarize_quality(threshold)
    
    dataT = torch.tensor(data[cols_to_z_score].values).float()
    labels = torch.tensor(data['bool_quality'].values).float()

    labels = labels[:, None]

    train_data, test_data, train_labels, test_labels = train_test_split(dataT, labels, test_size=.1)
    
    train_dataset = TensorDataset(train_data, train_labels)
    test_dataset = TensorDataset(test_data, test_labels)

    batch_size=32
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, drop_last=True)
    test_loader = DataLoader(test_dataset, batch_size=test_dataset.tensors[0].shape[0])
    print(f"Unbalanced dataset: {torch.unique(test_loader.dataset.tensors[1], return_counts=True)} ")
    
    return train_loader, test_loader
```

```python
class ANNWine(nn.Module):
    def __init__(self):
        super().__init__()
        
        self.input = nn.Linear(11, 16)
        
        self.fc1 = nn.Linear(16, 32)
        self.fc2 = nn.Linear(32, 32)
        
        self.relu1 = nn.LeakyReLU()
        self.relu2 = nn.LeakyReLU()
        self.relu3 = nn.LeakyReLU()
        
        self.output = nn.Linear(32, 1)
        
    def forward(self, x):
        
        x = self.relu1((self.input(x)))
        x = self.relu2((self.fc1(x)))
        x = self.relu3((self.fc2(x)))
          
        return self.output(x)
```

```python
# test model
model = ANNWine()
model(torch.randn(10, 11)).shape
```

```python
num_of_epochs = 500

def train_model(threshold):
    model = ANNWine()
    loss_func = nn.BCEWithLogitsLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=.001)
    
    train_loader, test_loader = get_train_test_data(threshold)
    
    train_acc = []
    test_acc = [] 
    losses = torch.zeros(num_of_epochs)
    
    for epoch in range(num_of_epochs):
        model.train()
        batch_acc = []
        batch_loss = []
        for X, y in train_loader:
            yHat = model(X)
            loss = loss_func(yHat, y)
            
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            predlabels = yHat>0
            batch_acc.append(100*torch.mean((predlabels==y).float()).item())
            batch_loss.append(loss.item())
            
        train_acc.append(np.mean(batch_acc))
        losses[epoch] = np.mean(batch_loss)
        
        model.eval()
        X,y = next(iter(test_loader))
        with torch.no_grad():
            predlabels = model(X)>0
        test_acc.append(100*torch.mean((predlabels==y).float()).item())
    
    return train_acc, test_acc, losses, model, train_loader, test_loader
```

```python
fig, ax = plt.subplots(3,3,figsize=(12,12))
thresholds = [4.5, 5.5, 6.5]
results = np.zeros((num_of_epochs, len(thresholds), 3))

for i, threshold in enumerate(thresholds):
    train_acc, test_acc, losses, model, train_loader, test_loader = train_model(threshold)
    results[:, i, 0] = losses
    results[:, i, 1] = train_acc
    results[:, i, 2] = test_acc
    
    ax[i, 0].plot(results[:, i, 0])
    ax[i, 0].set_xlabel('Epoch')
    ax[i, 0].set_ylabel('Loss')
    ax[i, 0].set_title(f'Losses with threshold {thresholds[i]}')
    
    ax[i, 1].plot(results[:, i, 1])
    ax[i, 1].plot(results[:, i, 2])
    ax[i, 1].set_xlabel('Epoch')
    ax[i, 1].set_ylabel('Accuracy (%)')
    ax[i, 1].set_ylim([10, 100])
    ax[i, 1].set_title(f'Accuracy with threshold {thresholds[i]}')   
    ax[i, 1].legend(['Train', 'Test'])
    
    X, y = next(iter(test_loader))
    
    predictions = model(X)
    item_accuracy = ((predictions > 0) == y).float()
    bad_wines_acc = 100*torch.mean(item_accuracy[y==0])
    good_wines_acc = 100*torch.mean(item_accuracy[y==1])

    
    bar = ax[i, 2].bar(['Bad', 'Good'], [bad_wines_acc, good_wines_acc])
    ax[i, 2].set_xlabel('Wine quality')
    ax[i, 2].set_ylabel('Test accuracy (%)')
    ax[i, 2].set_title(f'Per-qual Accuracy with threshold {thresholds[i]}')  
    
    for bar_idx, bar_item in enumerate(bar):
        N = torch.sum(train_loader.dataset.tensors[1]==bar_idx).item()
        ax[i, 2].text(bar_item.get_x()+bar_item.get_width()/2, bar_item.get_height()+1,
                     f"N={N}", ha='center', va='bottom', fontsize=14)
    
    
plt.tight_layout()
plt.show()
```
