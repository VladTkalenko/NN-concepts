```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torchvision

%matplotlib inline
import matplotlib_inline
matplotlib_inline.backend_inline.set_matplotlib_formats('svg')
```

```python
data_frame = pd.read_csv('./mnist_train.csv', delimiter=',')
```

```python
data_frame.head()
```

```python
data = data_frame.to_numpy()
```

```python
data.shape
```

```python
labels = data[:, 0]
data = data[:, 1:]

print(labels.shape)
print(data.shape)
```

```python
fig, axes = plt.subplots(3,4,figsize=(10,6))

for ax in axes.flatten():
    rand_img_2_show = np.random.randint(0, high=data.shape[0])
    
    img = np.reshape(data[rand_img_2_show, :], (28, 28))
    ax.imshow(img, cmap='gray')
    
    ax.set_title(f'The number {labels[rand_img_2_show]}')

plt.suptitle('How humans see the data', fontsize=20)
plt.tight_layout(rect=[0,0,1,.95])
plt.show()
```

```python
fig, axes = plt.subplots(3,4,figsize=(10,6))

for ax in axes.flatten():
    rand_img_2_show = np.random.randint(0, high=data.shape[0])
    
    ax.plot(data[rand_img_2_show, :], 'ko')
    ax.set_title(f'The number {labels[rand_img_2_show]}')
    
plt.suptitle('How the model sees the data', fontsize=20)
plt.tight_layout(rect=[0,0,1,.95])
plt.show()
```

```python
the7s = np.where(labels==7)[0]

fig, axs = plt.subplots(2,6, figsize=(15,6))

for i, ax in enumerate(axs.flatten()):
    img = np.reshape(data[the7s[i],:],(28,28))
    ax.imshow(img, cmap='gray')
    ax.axis('off')
    
plt.suptitle("Example 7's", fontsize=20)
plt.tight_layout(rect=[0,0,1,.95])
plt.show()
```

```python
print(data[the7s, :].shape)

C = np.corrcoef(data[the7s, :])

fig, ax = plt.subplots(1,3, figsize=(12,6))
ax[0].imshow(C, vmin=0, vmax=1)
ax[0].set_title("Correlation across all 7's")

uniqueCs = np.triu(C, k=1).flatten()
ax[1].hist(uniqueCs[uniqueCs!=0], bins=100)
ax[1].set_title('All unique correlations')
ax[1].set_xlabel("Correlations of 7's")
ax[1].set_ylabel('Count')

aveAll7s = np.reshape(np.mean(data[the7s, :], axis=0), (28,28))
ax[2].imshow(aveAll7s, cmap='gray')
ax[2].set_title("All 7's averaged together")

plt.tight_layout()
plt.show()
```
